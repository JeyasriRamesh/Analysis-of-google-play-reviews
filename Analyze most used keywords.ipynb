```python
import pandas as pd
import re
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.stem.wordnet import WordNetLemmatizer
from collections import Counter 
```

```python
stop_words = set(stopwords.words("english"))
new_words = ["app","company","card","time","use","also","even", "red",\
             "carpet", "redcarpet", "day", "still", "money","amount",\
             "month","pay","please","one","u","get","give","show","guy",\
             "say", "till", "ur", "try", "since", "ruby", "emi", "like", \
             "seen", "know", "every", "need", "already"]
stop_words = stop_words.union(new_words)
```

```python
df = pd.read_excel("Rating_5.xlsx")
df
```

```python
keyword_cloud=''
for i in range(len(df)):
    if df['Year'][i] == 2019:
        keyword_cloud = keyword_cloud + df['Review'][i]
    
print(len(keyword_cloud))

i=1
if (i==1):
    
    text= re.sub('[^a-zA-Z]', ' ', keyword_cloud)
    
    #Convert to lowercase
    text = text.lower()

    #remove tags
    text=re.sub("&lt;/?.*?&gt;"," &lt;&gt; ",text)

    # remove special characters and digits
    text=re.sub("(\\d|\\W)+"," ",text)

    ##Convert to list from string
    text = text.split()

    ##Stemming
    ps=PorterStemmer()
    #Lemmatisation
    lem = WordNetLemmatizer()
    text = [lem.lemmatize(word) for word in text if not word in  
            stop_words] 
    print(type(text))
    CounterVar = Counter(text)
    most_occur = CounterVar.most_common(50) 
    
print(most_occur)
```
